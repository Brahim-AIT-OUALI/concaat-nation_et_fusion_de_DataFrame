Concat - Premier exemple

```python
import pandas as pd
```

```python
df1= pd.DataFrame(
    {
        "Nom" : ["Julie", "Marc", "Julien"],
        "Age" : [25, 44, 33],
        "Equipe" : ["Argentine", "France", "Croatie"]
    }
    )


df2= pd.DataFrame(
    {
        "Nom" : ["Nathan", "Emma", "Thomas"],
        "Age" : [27, 51, 27],
        "Equipe" : ["Maroc", "Angleterre", "Pays-Bas"]
    }
)



df3= pd.DataFrame(
    {
        "Nom" : ["Martin", "Stanislas", "Marie"],
        "Age" : [2535, 42, 24],
        "Equipe" : ["Brésil", "Portugal", "Pologne"]
    }
    )
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
pd.concat([df1, df2, df3])
#On insère en premier lieu l'object
# Par défaut, la concaténation se faut par lignes.
```

Distinguer les différents morceaux du nouveau DF

```python
result = pd.concat([df1, df2, df3], keys = ["a", "b", "c"])
```

```python
result 
```

```python
result.loc["c"]
```

Réinialiser l'index

```python
result= pd.concat([df1, df2, df3], ignore_index =True)
#pd.concat(obj, ignore_index =False)
```

```python
result
```

```python
df4 = pd.DataFrame(
    {
        "Salaire": ["2980€", "241€", "917€", "1680€", "1975€"],
        "Tournoi": ["Champion", "Finale", "Troisième", "Demi-Finale", "Quart-de-finale"]    }
)
```

```python
# pd.concat(obj, axis = 0, join = "outer")
```

join = "outer" --- UNION
join = "inner" --- INTERSECTION

UNION - Aucune information n'est perdue
INTERSECTION - On supprime les lignes vides lors de la concaténation.

```python
pd.concat([result, df4], axis=1, join = "outer")
```

```python
pd.concat([result, df4], axis=1, join = "inner")
```

Merge - Premier exemple

```python
left = pd.DataFrame(
    {
        "Entreprise": ["A", "B", "C", "E", "I"],
        "Emploi": [1, 2, 3, 4, 5],
        "CA": [6, 7, 8, 9, 10],
        "Salarié": [11, 12, 13, 14, 15],
        "Lieu": ["Paris", "Madrid", "Londres", "Rome", "Berlin"]
            }
)

right = pd.DataFrame(
    {
        "Entreprise": ["A", "A", "C", "E"],
        "Secteur": [8, 9, 10, 11],
        "Site_Internet": [12, 13, 15, 17],
        "Lieu": ["Paris", "Madrid", "Londres", "Bruxelles"]
            }
)

```

```python
left
```

```python
right
```

```python
pd.merge(left, right)
# left - A DataFrame or Serie
# right - Another DataFrame or Serie
# Par défaut ça prendre les 2 colonnes communes. 
```

Fusionner sur une colonne

```python
pd.merge(left, right, on = "Entreprise")
```

Fusionner sur plusieurs colonne

```python
pd.merge(left, right, on = ["Entreprise", "Lieu"])
```

```python
# KeyError
pd.merge(left, right, on = ["Entreprise", "Secteur"])
```

La manière de fusionner

```python
# inner - default (intersection)- pd.merge(left, right, how = "inner")
# outer - union - aucune info n'est perdue
# left - prendre en compte toutes les lignes du left DataFrame + les valeurs communes du right DataFrame
# right - prendre en compte toutes les lignes du right DataFrame + les valeurs communes du left DataFrame
```

```python
pd.merge(left, right, how = "inner")
```

```python
pd.merge(left, right, how = "outer")
```

```python
pd.merge(left, right, how = "left")
```

```python
pd.merge(left, right, how = "right")
```

```python
pd.merge(left, right, how = "cross")
```

Fusionner 2 DataFrame avec 2 des noms de colonnes différents

```python
left = pd.DataFrame(
    {
        "Entreprise": ["A", "B", "C", "E", "I"],
        "Emploi": [1, 2, 3, 4, 5],
        "CA": [6, 7, 8, 9, 10],
        "Salarié": [11, 12, 13, 14, 15],
        "Lieu": ["Paris", "Madrid", "Londres", "Rome", "Berlin"]
            }
)

right = pd.DataFrame(
    {
        "Société": ["A", "A", "C", "E"],
        "Secteur": [8, 9, 10, 11],
        "Site_Internet": [12, 13, 15, 17],
        "Lieu": ["Paris", "Madrid", "Londres", "Bruxelles"]
            }
)

```

```python
# pd.merge(left, right, on = None, left_on = None, right_on=None)
pd.merge(left, right, left_on = "Entreprise", right_on="Société")
```

```python
# pd.merge(left, right, left_on = None, on = None, right_on=None, suffixes =("_x, _y"))
pd.merge(left, right, left_on = "Entreprise", right_on="Société", suffixes =("_l", "_r"))
```
